```python
def reset_seed(number):
    """
    Reset random seed to the specific number

    Inputs:
    - number: A seed number to use
    """
    random.seed(number)
    torch.manual_seed(number)
    return
```

```python
def tensor_to_image(data):
    """
    Convert a torch tensor into a numpy ndarray for visualization.

    Inputs:
    - tensor: A torch tensor of shape (1, H, W) with elements in the range [0, 1]

    Returns:
    - ndarr: A uint8 numpy array of shape (H, W, 1)
    """
    data = data.mul(255).clamp_(0, 255).permute(1, 2, 0)
    ndarr = data.to("cpu", torch.uint8).numpy()
    return ndarr
```

```python
def get_data(dataset_name, validation = True, validation_ratio = 0.2, show_examples = False, classes = None):
    """
    input:
    - dataset_name: name of dataset
    - validation: boolean, whether to split training dataset into training & validation
    - validation_ratio: ratio of training dataset for validation
    
    output:
    - train_set, [Optional] val_set(validaiton=True), test_set:  
    """
    dataset_name = dataset_name.lower()

    if dataset_name in ['speech']:
        train_set,test_set= get_dataset() # dataset part

    elif dataset_name in ['mnist','fashionmnist', 'cifar10','tinyimagenet']:
        datasets = load_datasets(dataset_name)
        X_train, y_train = datasets[:2]
        X_test, y_test = datasets[2:]
        print('Train data shape: ', X_train.shape)
        print('Train labels shape: ', y_train.shape)
        print('Test data shape: ', X_test.shape)
        print('Test labels shape: ', y_test.shape)
        
        # Visualize some examples from the dataset.
        if show_examples:
            if classes == None:
                classes = list(np.sort(np.unique(y_train)))
                
            samples_per_class = 10
            samples = []
            reset_seed(0)
            for y, cls in enumerate(classes):
                plt.text(-4, 30 * y + 18, cls, ha="right")
                (idxs,) = (y_train == y).nonzero()
                for i in range(samples_per_class):
                    idx = idxs[random.randrange(idxs.shape[0])].item()
                    samples.append(torch.tensor((X_train[idx]/255.0).reshape(1, 28, 28)))
            img = torchvision.utils.make_grid(samples, nrow=samples_per_class)
            plt.imshow(tensor_to_image(img))
            plt.axis("off")
            plt.show()

        
        transf_train, transf_test = data_transforms(dataset_name)
        if validation:
            num_training = int(X_train.shape[0] * (1.0 - validation_ratio))
            num_validation = X_train.shape[0] - num_training
            X_val = X_train[num_training : num_training + num_validation, :, :]
            y_val = y_train[num_training : num_training + num_validation]
            X_train = X_train[0:num_training, :, :]
            y_train = y_train[0:num_training]
            val_set = DatasetFromNumpy(X_val, y_val, transform=transf_train)
            
        train_set = DatasetFromNumpy(X_train, y_train, transform=transf_train)
        test_set = DatasetFromNumpy(X_test, y_test, transform=transf_test)
    if validation:
        return train_set, val_set, test_set
    else:
        return train_set, test_set
```

# Define CNN Network

```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
class FashionCNN(nn.Module):
    '''
    CNN for Fashion MNIST 
    
    Args:
        bias : Boolean, whether to add bias 
    
    Returns:
        object : initialised model
    '''
    def __init__(self, bias = False):
        super().__init__()
#         self.conv1 = nn.Conv2d(1, 32, 5, stride = 1, padding = 0, bias=bias)
#         self.conv2 = nn.Conv2d(32, 64, 3, stride = 1, padding = 0, bias=bias)
#         self.conv3 = nn.Conv2d(64, 128, 3, stride = 1, padding = 0, bias=bias)
#         self.fc1 = nn.Linear(1152, 128, bias=bias)
#         self.fc2 = nn.Linear(128, 10, bias=bias)

        self.conv1 = nn.Conv2d(1, 32, 5, stride = 1, padding = 2, bias=bias)
        self.conv2 = nn.Conv2d(32, 32, 3, stride = 1, padding = 1, bias=bias)
        self.conv3 = nn.Conv2d(32, 64, 3, stride = 1, padding = 1, bias=bias)
        self.fc1 = nn.Linear(576, 10, bias=bias)
        
#         self.conv1 = nn.Conv2d(1, 16, 5, stride = 1, padding = 0, bias=bias)
#         self.conv2 = nn.Conv2d(16, 16, 3, stride = 1, padding = 0, bias=bias)
#         self.conv3 = nn.Conv2d(16, 32, 3, stride = 1, padding = 0, bias=bias)
#         self.conv4 = nn.Conv2d(32, 32, 3, stride = 1, padding = 0, bias=bias)
#         self.conv5 = nn.Conv2d(32, 64, 3, stride = 1, padding = 0, bias=bias)
#         self.fc1 = nn.Linear(576, 10, bias=bias)
        
    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.max_pool2d(x, 2)
        x = F.relu(self.conv2(x))
        x = F.max_pool2d(x, 2)
        x = F.relu(self.conv3(x))
        x = F.max_pool2d(x, 2)
        x = x.view(x.size(0), -1)
        out = self.fc1(x)
        return out
```

```python
def eval_model(model, device, testloader, loss_fn=None):
    
    '''
    Function which evaluates the neural network model
    
    Args:
    
        model : model for evaluating
        device : cpu or gpu
        test_loader: the testing set data loader
        loss_fn : criterion to optimize on (loss function)
    
    Returns:
        mean_loss, accuracy : Returns mean loss and accuracy of the model on testing dataset 
    '''
    
    model.eval()
    correct = 0
    total = 0
    losses = []
    with torch.no_grad():
        for i, (images, labels) in enumerate(testloader):
            images = images.to(device)
            labels = labels.to(device)
            
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum()
            
            if loss_fn is not None:
                loss_eval = loss_fn(outputs, labels)
                losses.append(loss_eval.item())
                
    if not loss_fn:
        mean_loss = -1
    else:
        mean_loss = np.mean(losses)
        
    accuracy = correct.cpu().numpy()/total
    return mean_loss, accuracy
```

```python
## Define training parameters

EPOCHS = 50 # number of training epochs
LR = 1e-3 # learning rate
WEIGHT_DECAY = 2e-2 # weight decay rate
SAVE = True # whether to save 
model_name = "fashion_cnn_FP"
```

```python
model = FashionCNN(bias=False)
model.to(device)
```

```python
_, acc = eval_model(model.to(device), device, testloader)
print('Init model accuracy:', acc)
acc_run = acc
best_val_loss = np.inf
best_acc = 0

loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5)


history = {name:[] for name in ['train', 'train_accuracy', 'val', 'val_accuracy', 'lr']}

for epoch in range(EPOCHS):
    model.train()
   
    for i, (images, labels) in enumerate(trainloader):
        
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        
        pred = model(images)
        loss_t = loss_fn(pred, labels)
        
        loss = loss_t
        loss.backward()

        optimizer.step()
        
    train_loss, train_acc = eval_model(model, device, trainloader, loss_fn)
    val_loss, val_acc = eval_model(model, device, validationloader, loss_fn)
    acc_run = max(acc_run, val_acc)
    
    scheduler.step(val_loss)
    
    history['train'].append(train_loss)
    history['val'].append(val_loss)
    history['train_accuracy'].append(train_acc)
    history['val_accuracy'].append(val_acc)
    history['lr'].append(optimizer.param_groups[0]['lr'])
   
    print(f'Epoch [{epoch+1}/{EPOCHS}] | Train_Loss: {train_loss:.4f} | Val_loss: {val_loss:.4f} | Train_accuracy: {train_acc:.4f}| Val_accuracy: {val_acc:.4f}')

    is_best = val_loss <= best_val_loss

    if is_best:
        best_st_dict = model.state_dict() 
        best_val_loss = val_loss
        
if SAVE:
    hist = pd.DataFrame(history)
    hist.to_csv("../saved_models/" + f'{model_name}.csv', index=False)
    
    torch.save(best_st_dict, "../saved_models/" + f'{model_name}.pt')
```

```python
def plot_training_history(data, title, x_label = 'epochs', label_marker = 'o-', labels = None):
    """utility func for plotting training history"""
    plt.title(title)
    plt.xlabel(x_label)
    for i in range(len(data)):
        if labels is not None:
            plt.plot(data[i], label_marker, label = labels[i])
        else:
            plt.plot(data[i], label_marker)
    plt.legend(loc = 'lower right')
```

```python
def default_Arg_CB(args, gp, weight_bit, noisy=0, sigma=1.0, prog_sigma = 1.0, ADC_sigma=2.0, onoff_ratio_ideal=False, onoff_ratio=10, w_max = 3.0, mode = "training"):
    args.weight_bits = weight_bit # weight bit of the model, not the device
    args.noisy = noisy
    args.mode = mode
    args.training_mode = 'fixed_w_full_range_CB_quant'

    args.QInpFlag=True
    args.QWeightFlag=True
    args.QActFlag=True
    
    args.tile = True
    args.dual_array = True
    args.set_inp_scale = True
    args.set_act_scale = True
    
    gp.weight_bits = args.weight_bits - 1
    gp.weight_prog_bits = args.weight_bits - 1
    gp.onoff_ratio_ideal = onoff_ratio_ideal
    gp.onoff_ratio = onoff_ratio
    gp.cell_I_max = w_max
    gp.cell_I_min = w_max/gp.onoff_ratio
    
    args.inp_quant_type = 'inp_dynamic'
    args.act_quant_type = 'ADC_dynamic'

    if not noisy:
        gp.if_weight_noise = False
        gp.weight_noise_type = None
        gp.if_ADC_noise = False
        gp.if_weight_prog_noise = False
        gp.weight_noise_sigma = 0
        gp.ADC_noise_sigma = 0
        gp.weight_prog_noise_sigma = 0
    else:
        gp.if_weight_noise = True
        gp.weight_noise_type = 'sigma_relative'
        gp.weight_noise_sigma = sigma
        gp.if_ADC_noise = True
        gp.ADC_noise_type = 'sigma_relative'
        gp.ADC_noise_sigma = ADC_sigma
        gp.if_weight_prog_noise = True
        gp.weight_prog_noise_sigma = prog_sigma
    gp.update()
    gp.summary()
    return args
```

```python
model_params = []
weight_params = []
weight_labels = []

cnt_conv = 1
cnt_fc = 1
for m in model_CB.modules():
    if isinstance(m, Tile_Conv2d):
        print("cnt_conv", cnt_conv)
        if m.Flag_bias:
            model_params.append(m.bias)
        if cnt_conv == 1:
            if m.set_act_scale:
                model_params.append(m.act_scaling)
            for layer in range(m.n_rs):
                weight_params.append(m.tile_module[layer].dual_module[0].weight)
                weight_labels.append('Conv{}_{}'.format(cnt_conv, layer))
        else:
            model_params.append(m.uI)
            if m.set_inp_scale:
                model_params.append(m.inp_scaling)
            if m.set_act_scale:
                model_params.append(m.act_scaling)
            for layer in range(m.n_rs):
                weight_params.append(m.tile_module[layer].dual_module[0].weight)
                weight_labels.append('Conv{}_{}'.format(cnt_conv, layer))
        cnt_conv += 1
    elif isinstance(m, Tile_Linear):
        if m.Flag_bias:
            model_params.append(m.bias)
        model_params.append(m.uI)
        if m.set_inp_scale:
            model_params.append(m.inp_scaling)
        if m.set_act_scale:
            model_params.append(m.act_scaling)
            for layer in range(m.n_rs):
                weight_params.append(m.tile_module[layer].dual_module[0].weight)
                weight_labels.append('FC{}_{}'.format(cnt_fc, layer))
        cnt_fc += 1
    elif isinstance(m, Dual_Conv2d) or isinstance(m, Dual_Linear):
        if m.Flag_bias:
            model_params.append(m.bias)
    elif isinstance(m, nn.BatchNorm2d) or isinstance(m, nn.BatchNorm1d):
        if m.affine:
            model_params.append(m.weight)
            model_params.append(m.bias)
lw = len(weight_params)
print(lw)
print(len(model_params))
```

```python
EPOCHS = 50 
SAVE = True
acc_run = acc
best_val_loss = np.inf
best_acc = 0

loss_fn = nn.CrossEntropyLoss()
optimizer_m = torch.optim.AdamW(model_params, lr=args.lr_m, weight_decay=args.weight_decay_w)
scheduler_m = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer_m, mode='min', factor=0.5, patience=5)

optimizer_w = torch.optim.AdamW(weight_params, lr=args.lr_w, weight_decay=args.weight_decay_m)
scheduler_w = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer_w, mode='min', factor=0.5, patience=5)


history = {name:[] for name in ['train', 'train_accuracy', 'val', 'val_accuracy', 'lr_w', 'lr_m']}

for epoch in range(EPOCHS):
    model_CB.train()
   
    for i, (images, labels) in enumerate(trainloader):
        
        images = images.to(device)
        labels = labels.to(device)

        optimizer_w.zero_grad()
        optimizer_m.zero_grad()
        
        pred = model_CB(images)
        loss_t = loss_fn(pred, labels)
        
        loss = loss_t
        loss.backward()

        optimizer_w.step()
        optimizer_m.step()
        
    train_loss, train_acc = eval_model(model_CB, device, trainloader, loss_fn)
    val_loss, val_acc = eval_model(model_CB, device, validationloader, loss_fn)
    acc_run = max(acc_run, val_acc)
    
    scheduler_w.step(val_loss)
    scheduler_m.step(val_loss)
    
    history['train'].append(train_loss)
    history['val'].append(val_loss)
    history['train_accuracy'].append(train_acc)
    history['val_accuracy'].append(val_acc)
    history['lr_w'].append(optimizer_w.param_groups[0]['lr'])
    history['lr_m'].append(optimizer_m.param_groups[0]['lr'])
   
    print(f'Epoch [{epoch+1}/{EPOCHS}] | Train_Loss: {train_loss:.4f} | Val_loss: {val_loss:.4f} | Train_accuracy: {train_acc:.4f}| Val_accuracy: {val_acc:.4f}')

    is_best = val_loss <= best_val_loss

    if is_best:
        best_st_dict = model_CB.state_dict() 
        best_val_loss = val_loss
        
if SAVE:
    hist = pd.DataFrame(history)
    hist.to_csv("../saved_models/" + f'{model_name}.csv', index=False)
    
    torch.save(best_st_dict, "../saved_models/" + f'{model_name}.pt')
    
plt.subplot(2, 1, 1)
plot_training_history([history["train"], history['val']], "Loss", labels = ['Train_loss', 'Val_loss'])

plt.subplot(2, 1, 2)
plot_training_history([history["train_accuracy"], history['val_accuracy']], "Loss", labels = ['Train_accuracy', 'Val_accuracy'])

plt.title(f"Quant-aware training: RRAM {gp.weight_bits}bits")
plt.gcf().set_size_inches(7, 12)
plt.tight_layout()
plt.show()
```
